## Question 1

From ChagGPT

For a binary classification problem, the most appropriate loss function is B) binary crossentropy.

Explanation:

Mean Squared Error (MSE): This loss function is commonly used for regression problems where the output is a continuous value. In binary classification, however, we are dealing with a categorical problem, not a regression problem.

Binary Crossentropy: This is the standard loss function for binary classification problems. It measures the difference between the true distribution and the predicted distribution. It works well when the output of your model is a probability score between 0 and 1, representing the likelihood of the example belonging to the positive class. Binary crossentropy is suitable for problems where each example belongs to exactly one of two classes.

Categorical Crossentropy: This is used when dealing with multi-class classification problems, not binary. It assumes that each example belongs to one and only one class out of multiple classes.

Cosine Similarity: This is a metric used for measuring the cosine of the angle between two non-zero vectors. It's often used for similarity between two continuous vectors but not as a loss function for binary classification problems.

Therefore, in a binary classification problem, binary crossentropy is the most appropriate choice as it is designed to work with binary outcomes and is effective for optimizing models in this context.

## Package installation

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
!pip3 install tensorflow
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.preprocessing.image import load_img
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.optimizers import SGD
```

## Compile the model as described

```python
# Define the CNN model
model = Sequential()

# Convolutional layer with 32 filters, kernel size (3, 3), and 'relu' activation
model.add(Conv2D(32, (3, 3), input_shape=(150, 150, 3), activation='relu'))

# MaxPooling layer with pool size (2, 2)
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten layer to convert the result into vectors
model.add(Flatten())

# Dense layer with 64 neurons and 'relu' activation
model.add(Dense(64, activation='relu'))

# Output layer with 1 neuron and appropriate activation for binary classification
model.add(Dense(1, activation='sigmoid'))

# Compile the model with SGD optimizer and specified parameters
sgd_optimizer = SGD(lr=0.002, momentum=0.8)
model.compile(optimizer=sgd_optimizer, loss='binary_crossentropy', metrics=['accuracy'])
```

## Question 2

```python
# Display the model summary
model.summary()
```

## Question 3

```python
# Import ImageDataGenerator
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Build image generator for training (takes preprocessing input function)
train_gen = ImageDataGenerator(rescale=1./255)

# Load in train dataset into train generator
train_ds = train_gen.flow_from_directory(directory=r'./hm8/data/train',
                                         target_size=(150,150),
                                         batch_size=20,
                                         class_mode='binary',
                                         shuffle=True)

# Create image generator for validation
val_gen = ImageDataGenerator(rescale=1./255)

# Load in image for validation
val_ds = val_gen.flow_from_directory(directory=r'./hm8/data/test', # Validation image directory
                                     target_size=(150,150),
                                     batch_size=20,
                                     class_mode='binary',
                                     shuffle=True)
```

```python
history = model.fit(
    train_ds,
    epochs=10,
    validation_data=val_ds
)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and Validation Accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and Validation Loss')
plt.legend()

plt.show()
```

```python
acc_median = np.median(acc)
acc_median
```

## Question 4

```python
loss_std = np.std(loss)
loss_std
```
